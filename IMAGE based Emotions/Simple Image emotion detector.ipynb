# Simple Image emotion detector(using jpg,jpeg files)

```python
import cv2
import matplotlib.pyplot as plt
import os
import warnings
warnings.filterwarnings('ignore')
```

# Load the file

```python
img = cv2.imread('angry2.jpg')
print(img)
```

# Plot the img

```python
plt.imshow(img)
```

# Plot using Red,Green,Blue(RGB)

```python
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
```

# Using deepface we can predict

```python
from deepface import DeepFace
predictions = DeepFace.analyze(img)
```

```python
predictions
```

```python
result = DeepFace.analyze(img, actions = ['emotion']) 
```

```python
print("Predicted emotion is:",result['dominant_emotion'])
```

# Use haarcascade_frontalface

```python
import cv2
from deepface import DeepFace
faceCascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
#print(faceCascade.empty())
faces = faceCascade.detectMultiScale(gray,1.1,4)
# Draw a rectangle around the faces
for (x, y, w, h) in faces:
         cv2.rectangle(img, (x, y), (x+w, y+h), (0, 255, 0), 2)
```

```python
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
```

```python
font =cv2.FONT_HERSHEY_SIMPLEX
     # Use putText() method for
     # inserting text on video
cv2.putText(img,
            predictions['dominant_emotion'],
            (0, 50),
            font, 1,
            (0, 0, 255),
            2,
            cv2.LINE_4) ;
```
